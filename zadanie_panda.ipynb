```python
import pandas as pd

prices = [
	(1, 2.12),
	(2, 2.56),
	(3, 3.10),
	(4, 3.16),
	(5, 3.58),
	(6, 5.12),
	(7, 5.16),
	(8, 5.20),
	(9, 4.12),
	(10, 4.10),
	(11, 3.65),
	(12, 4.25),

]
df = pd.DataFrame(prices)

df = pd.DataFrame(prices, columns=["month", "PLN"])
df = df.set_index("month")

# Declare a list that is to be converted into a column 
usd=[]
  
for item in range(12):
    
    usd.append(prices[item][1]/4)
   
   
df['USD'] = usd    

  
df["USD"].plot(color = ['#BB0000'],linestyle='dashed', title='Price of googds [USD]')
```
